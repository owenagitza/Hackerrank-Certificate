## **Alphabet Rangoli**

**Original**

````python
def print_rangoli(size):
    
    # Set alphabet
    alphabet = 'abcdefghijklmnopqrstuvwxyz'

    n = size

    # Width
    w = (4 * n) - 3

    # Upper + middle
    for i in range(n):
        print(('-'.join(alphabet[n - 1:n - (i + 1):-1] + alphabet[n - (i + 1):n])).center(w, '-'))
        
    # Lower
    for i in range(1, n):
        print(('-'.join(alphabet[n - 1:i:-1] + alphabet[i:n])).center(w, '-'))

if __name__ == '__main__':
    n = int(input())
    print_rangoli(n)
````

````python
def merge_the_tools(string, k):
    # Parsing the string
    l = [string[i * k:(i + 1) * k] for i in range(len(string) // k)]

    # Check for any repeated string
    for substring in l:
        result = ''
        used_char = []
        for i in substring:
            if i not in used_char:
                result += i
                used_char.append(i)
            else:
                continue
        print(result)

if __name__ == '__main__':
    string, k = input(), int(input())
    merge_the_tools(string, k)
````

`itertools.groupby()`
```
itertools.groupby(iterable, key = None)

````python
import itertools

l = [0, 0, 0, 1, 1, 2, 0, 0]

for k, g in itertools.groupby(l):
    print(k, g)
# 0 <itertools._grouper object at 0x00000216D3795300>
# 1 <itertools._grouper object at 0x00000216D3795780>
# 2 <itertools._grouper object at 0x00000216D3795300>
# 0 <itertools._grouper object at 0x00000216D3795780>
    
for k, g in itertools.groupby(l):
    print(k, list(g))
# 0 [0, 0, 0]
# 1 [1, 1]
# 2 [2]
# 0 [0, 0]
````

````python
print([k for k, g in itertools.groupby(l)])
# [0, 1, 2, 0]

print([list(g) for k, g in itertools.groupby(l)])
# [[0, 0, 0], [1, 1], [2], [0, 0]]

print([(k, list(g)) for k, g in itertools.groupby(l)])
# [(0, [0, 0, 0]), (1, [1, 1]), (2, [2]), (0, [0, 0])]
````

**Remarks**

- The Python `defaultdict` type behaves almost exactly like a regular Python dictionary that's specifically designed to handle **missing keys**
- But if you try to access or modify a missing key, then `defaultdict` will automatically create the key and generate a default value for it
- This makes `defaultdict` a valuable option for handling missing keys in dictionaries

`collections.defaultdict()`
```
collections.Counter(default_factory = None)

````python
from collections import defaultdict

# Correct instantiation
dd = defaultdict(list) # Pass list to default_factory
dd['one'] = 1 # Add a key-value pair
print(dd['missing']) # Access a missing key returns an empty list
dd['another_missing'].append(4) # Modify a missing key
print(dd)
````

- Remember not to call it using the parentheses (`defaultdict(list)` not `defaultdict(list())`)

## **group(), groups() & groupdict()**

**Remarks**

`group()`
```
Match.group([group1, ...])
```
- Returns one or more subgroups of the match. If there is a single argument, the result is a single string; if there are multiple arguments, the result is a tuple with one item per argument
- If a groupN argument is zero, the corresponding return value is the entire matching string

````python
import re

m = re.match(r'(\w+)@(\w+)\.(\w+)', 'username@hackerrank.com')
m.group(0)       # The entire match
'username@hackerrank.com'
m.group(1)       # The first parenthesized subgroup
'username'
m.group(2)       # The second parenthesized subgroup
'hackerrank'
m.group(3)       # The third parenthesized subgroup
'com'
m.group(1,2,3)   # Multiple arguments give us a tuple
# ('username', 'hackerrank', 'com')
````

`groups()`
```
Match.groups(default = None)
```
- Return a tuple containing all the subgroups of the match, from 1 up to however many groups are in the pattern
- The default argument is used for groups that did not participate in the match; it defaults to `None`

````python
import re

m = re.match(r'(\w+)@(\w+)\.(\w+)','username@hackerrank.com')
m.groups()
# ('username', 'hackerrank', 'com')
````

`groupdict()`
```
Match.groupdict(default = None)
```
- Return a dictionary containing all the named subgroups of the match, keyed by the subgroup name
- The default argument is used for groups that did not participate in the match; it defaults to `None`

````python
import re

m = re.match(r'(?P<user>\w+)@(?P<website>\w+)\.(?P<extension>\w+)','myname@hackerrank.com')
m.groupdict()
# {'website': 'hackerrank', 'user': 'myname', 'extension': 'com'}
````

## **Regex Substitution**

**Original**

````python
import re

def substitute(match):
    symb = match.group()
    if symb == '&&':
        return 'and'
    else:
        return 'or'

n = int(input())
pattern = re.compile(r'(?<= )(&&|\|\|)(?= )')
l = [input() for _ in range(n)]
for s in l:
    subbed = pattern.sub(substitute, s)
    print(subbed)
````

**Remarks**

`sub()`
```
re.sub(pattern, repl, string, count = 0, flags = 0)
```
- Returns the string obtained by replacing the leftmost non-overlapping occurrences of `pattern` in `string` by the replacement `repl`
- If the `pattern` isnt found, `string` is returned unchanged
- `repl` can be a string or a function; if it is a string, any backslash escapes in it are processed. That is, `\n` is converted to a single newline character
- If it is a function, then the match object is passed unto the function
- Backreferences, such as \6, are replaced with the substring matched by group 6 in the pattern

````python
import re

#Squaring numbers
def square(match):
    number = int(match.group(0))
    return str(number ** 2)

print(re.sub(r"\d+", square, "1 2 3 4 5 6 7 8 9"))
````

***
- Here are the key points to solve this problem
- `re.sub()` can be used to pass a `Match` object unto a function
- It should accomodate overlapping cases, hence it should only match a word block, i.e. separated by whitespaces
- To achieve this, I tried using word boundaries `\b` but it didn't work
- Hence, I used lookarounds
```
    (?<= )(&&|\|\|)(?= )
```
- The meaning is:
    - Asserts that what immediately precedes the current position in the string is whitespace' ', followed by either '&&' or '||' and asserts that what immediately follows the current position in the string is whitespace ' '

````python
from html.parser import HTMLParser

class MyHTMLParser(HTMLParser):
    def handle_starttag(self, tag, attrs):
        print(f'Start : {tag}')
        for i in attrs:
            print(f'-> {i[0]} > {i[1]}')
        
    def handle_endtag(self, tag):
        print(f'End   : {tag}')
        
    def handle_startendtag(self, tag, attrs):
        print(f'Empty : {tag}')
        for i in attrs:
            print(f'-> {i[0]} > {i[1]}')
            
n = int(input())
parser = MyHTMLParser()
for _ in range(n):
    parser.feed(input())
````

## **Validating UID**

**Original**

````python
import re

def uid_validation(s):
    if bool(re.match(r'^[a-zA-Z0-9]{10}$', s)) == False: # Exactly 10 alphanumeric chars (Cond. 5 & 3)
        return print('Invalid')
    if len(s) > len(set(s)): # Checks for any duplication. If len(s) > len(set(s)) => There is duplication (Cond. 4)
        return print('Invalid')
    if len(re.sub(r'[^A-Z]', '', s)) < 2: # Checks for at least 2 uppercase English alphabet using len by first removing the negation of such condition by re.sub
        return print('Invalid')
    if len(re.sub(r'[^0-9]', '', s)) < 3: # Checks for at least 3 digits using len by first removing the negation of such condition by re.sub
        return print('Invalid')
    return print('Valid')
        
n = int(input())
for _ in range(n):
    uid_validation(input())
````

## **Validating Credit Card Numbers**

**Original**

````python
import re

def cc_validation(s):
    if bool(re.match(r'((^[456]\d{15}$)|(^[456]\d{3}-\d{4}-\d{4}-\d{4}$))', s)) == False: # Starts with 4, 5, or 6 and exactly 16 digits and consist only digits (Cond. 1, 2, 3)
        return print('Invalid')
    if bool(re.search(r'(\d)\1{3}|(\d)-\2{3}|(\d)\3\-\3{2}|(\d)\4{2}-\4', s)):
        return print('Invalid')
    return print('Valid')

n = int(input())
for _ in range(n):
    cc_validation(input())
````

## **XML2 - Find the Maximum Depth**

**Original**

````python
import xml.etree.ElementTree as etree

maxdepth = 0
def depth(elem, level):
    global maxdepth
    if (len(elem.findall('.//')) > 0):
        level += 1
        if (level >= maxdepth):
            maxdepth = level + 1
        for i in elem:
            depth(i, level)

if __name__ == '__main__':
    n = int(input())
    xml = ""
    for i in range(n):
        xml =  xml + input() + "\n"
    tree = etree.ElementTree(etree.fromstring(xml))
    depth(tree.getroot(), -1)
    print(maxdepth)
````

## **Standardize Mobile Number Using Decorators**

**Original**

````python
import re

# Defining decorator
def wrapper(function):
    def standardize(l):
        standardized_list = []
        for s in l:
            if len(s) == 10:
                pattern = re.compile(r'(\d{5})(\d{5})')
                replacement = r'+91 \1 \2'
                s = re.sub(pattern, replacement, s)
            else:
                pattern  = re.compile(r'^(0|91|\+91)(\d{5})(\d{5})')
                replacement = r'+91 \2 \3'
                s = re.sub(pattern, replacement, s)
            standardized_list.append(s)
        return function(standardized_list)
    return standardize

@wrapper
def sort_phone(l):
    print(*sorted(l), sep = '\n')

if __name__ == '__main__':
    l = [input() for _ in range(int(input()))]
    sort_phone(l) 
````

**Remarks**

The typical pattern for using decorators is:
```
# Decorator definition
def decorator_function(f):
    def wrapper_function():
        return f()
    return wrapper_function

# Usage
@decorator_function
function()
```

````python
import operator

# Defining decorator
def person_lister(function):
    def age_sorter(people):
        people.sort(key = lambda x: int(operator.itemgetter(2)(x)))
        sorted_list = []
        for person in people:
            sorted_list.append(function(person))
        return sorted_list
    return age_sorter

@person_lister
def name_format(person):
    return ("Mr. " if person[3] == "M" else "Ms. ") + person[0] + " " + person[1]

if __name__ == '__main__':
    people = [input().split() for i in range(int(input()))]
    print(*name_format(people), sep = '\n')
````

- The `name_format` function takes a list `(['Mike', 'Thomson', '20', 'M'])` and returns a formated string from a list `(Mr. Mike Thomson)`
- Since the final command prints out every items in the list, then the decorator should return a list containing formated names
- In the decorator definition, the wrapper function should take exactly the same input as the input function

## **Default Arguments**

**Original**

````python
class EvenStream(object):
    def __init__(self):
        self.current = 0

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

class OddStream(object):
    def __init__(self):
        self.current = 1

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

def print_from_stream(n, stream = None):
    if stream == None:
        stream = EvenStream()
    for _ in range(n):
        print(stream.get_next())

queries = int(input())
for _ in range(queries):
    stream_name, n = input().split()
    n = int(n)
    if stream_name == "even":
        print_from_stream(n)
    else:
        print_from_stream(n, OddStream())
````

**Remarks**

- The bug in this program is demonstrated as the following:

````python
class EvenStream(object):
    def __init__(self):
        self.current = 0

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

class OddStream(object):
    def __init__(self):
        self.current = 1

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

def print_from_stream(n, stream = EvenStream()):
    for _ in range(n):
        print(stream.get_next())

inputs = [['even', '2'], ['odd', '2'], ['even', '3']]
for l in inputs: 
    stream_name, n = l
    n = int(n)
    if stream_name == "even":
        print_from_stream(n)
    else:
        print_from_stream(n, OddStream())
````

- Notice that the output for `'even'` does not reset to `0` as expected
- Instead, it continues from where it left off (The last `'even'` input) at `2`
- This bug is caused by the default argument given to the `print_from_stream` function
- This is a core issue is that in Python, the value of a default argument is evaluated only once, when the function is declared. ***Its the same object every time the function is called***. For further explanation check [https://towardsdatascience.com/python-pitfall-mutable-default-arguments-9385e8265422]
***
- A common way to combat this bug is to use an immutable placeholder value like `None` for the default argument. The function looks for that placeholder value and, when found, replaces it with a brand new instance of the mutable type
- Look at the demonstration below

````python
class EvenStream(object):
    def __init__(self):
        self.current = 0

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

class OddStream(object):
    def __init__(self):
        self.current = 1

    def get_next(self):
        to_return = self.current
        self.current += 2
        return to_return

def print_from_stream(n, stream = None):
    if stream == None:
        stream = EvenStream()
    for _ in range(n):
        print(stream.get_next())

inputs = [['even', '2'], ['odd', '2'], ['even', '3']]
for l in inputs: 
    stream_name, n = l
    n = int(n)
    if stream_name == "even":
        print_from_stream(n)
    else:
        print_from_stream(n, OddStream())
````
